```python
import pandas as pd
import numpy as np
```

```python
#read the files
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
#show dataframe's shapes
print(file1.shape)
print(file2.shape)
file3.shape
```

```python
col1 = list(map(lambda el : el.lower(),file1))
file1.columns = col1
col2 = list(map(lambda el : el.lower(), file2))
file2.columns = col2
col3 = list(map(lambda el : el.lower(),file3))
file3.columns = col3
```

```python
#standardize header names
file1 = file1.rename(columns={'st':'state'})
file2 = file2.rename(columns={'st':'state'})
file1
```

```python
#remove file1 %
file1_custom = file1["customer lifetime value"].str.replace("%","")
file1['customer lifetime value'] = file1_custom
file1
```

```python
#remove file2 %
file2_custom = file2["customer lifetime value"].str.replace("%","")
file2['customer lifetime value'] = file2_custom
file2
```

```python
#rearrange file2 columns
file2_rearranged = file2[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file2_rearranged
```

```python
#concatenate dataframes
dataset0 = pd.concat([file1,file2],axis=0)
dataset0.shape
```

```python
#change dtype to numerical
new_column0 = pd.to_numeric(dataset0['customer lifetime value'], errors = 'coerce')
dataset0
dataset0['customer lifetime value'] = new_column0
dataset0.dtypes
dataset0 ['customer lifetime value'] = dataset0 ['customer lifetime value']/100
dataset0
```

```python
#rearrange file3 columns
file3_rearranged = file3[['customer', 'state', 'gender', 'education', 'customer lifetime value',
       'income', 'monthly premium auto', 'number of open complaints',
       'policy type', 'vehicle class', 'total claim amount']]
file3_rearranged.head()
```

```python
#concatenate dataframes
dataset = pd.concat([dataset0,file3],axis=0)
dataset.shape
```

```python
#check for categorical and numerical columns
dataset.dtypes
```

```python
dataset
```

```python
# drop columns
dataset_drops = dataset.drop(['education','number of open complaints'], axis=1)
dataset_drops.head()
```

```python
#divide customer lifetime value / 100
dataset_drops ['customer lifetime value'] = dataset_drops ['customer lifetime value']/100
dataset_drops
```

```python
#drop duplicates
dataset_drops = dataset_drops.drop_duplicates()
dataset_drops = dataset_drops.reset_index(drop=True)
dataset_drops
```

```python
#income >0
filtered_data = dataset_drops[(dataset_drops['income']>0)]
filtered_data
```
